## Task 1.2
population more than 10_000_000 & sorted by population
Clue: sorted()
Expected Output
['India', 'China', 'Russia'....]

````python
def population_countries(countries):
    population_countries = []
    for country in countries:
        if country["population"] >= 10_000_000:
            population_countries.append(
                {"name": country["name"]["common"], "population": country["population"]}
            )
    population_countries.sort(key=lambda country: country["population"], reverse=True)
    sorted_countries = [country["name"] for country in population_countries]

    return sorted_countries
````

# Task 2 (Challenging)
Create a function which takes list of cities and provide the corresponding temperature in a dictionary
 
```python
cities = ["Cape town", "Johannesburg", "Durban"]
```
 
## Expected Output
```python
{
"Cape town": 24,
"Johannesburg": 25,
"Durban": 18
}
```

````python
import requests

API_Key = "1e66fa84b3544c8283684358252403&q"

def get_city_temp(city_name):
    API = f"http://api.weatherapi.com/v1/current.json?key={API_Key}={city_name}&aqi=no"  # Replace with your OpenWeatherMap API key
    response = requests.get(API, verify=False)
    city_weather = response.json()
    return city_weather['location']['name'], city_weather['current']['temp_c']
cities = ["Cape town", "Johannesburg", "Durban"]

def get_temps(cities):
    city_temperatures = {}

    for city in cities:
        city_name, temp = get_city_temp(city)
        city_temperatures[city_name] = temps

    return city_temperatures
temps = get_temps(cities)
print(temps)
````
